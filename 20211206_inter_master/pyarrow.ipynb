```python
import time
```

```python
from pyarrow import csv
```

```python
def load_by_arrow(file):
    table=csv.read_csv(file)
    return table.to_pandas()
```

```python
import pandas as pd
```

```python
def load_by_pandas(file):
    df=pd.read_csv(file)
    return df
```

```python
%%time

file='goog_data.csv'

pa_df=load_by_arrow(file)
pa_df
```

```python
%%time

file='goog_data.csv'

df=load_by_pandas(file)
df
```
