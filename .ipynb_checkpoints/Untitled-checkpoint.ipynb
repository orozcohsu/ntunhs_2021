```python
import pandas as pd
```

```python
data={"a":[1,2,3,4,5],
      "b":[2,3,4,5,6],
      "c":[1,1,2,2,1]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
grouped = df.groupby('c').sum().reset_index()
grouped
```

```python
grouped.sort_values('b', ascending=True)
```

```python
grouped = df.groupby('c', sort=False).sum()
grouped
```

```python
grouped.sort_index(ascending=False)
```

```python
df.groupby('c')['a'].count().sort_values(ascending=False)
```

```python
import numpy as np
ex = pd.DataFrame(np.random.randint(1,10,size=(100,3)), columns=['features', 'AUC', 'recall'])
ex
```

```python

# pass a tuple corresponding to which specific col you want sorted. In this case, 'mean' or 'AUC' alone are not unique. 
ex.groupby('features').agg(['mean','std']).sort_values(('AUC', 'mean'))
```
