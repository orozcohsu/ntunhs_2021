```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
data = pd.DataFrame(diabetes.data,
                    columns=diabetes.feature_names)
data['diabetes'] = diabetes.target
```

```python
data
```

```python
print(diabetes.DESCR)
```

```python
#Interaction Features
data['s1+s2']=data['s1'].add(data['s2'])
data
```

```python
data['s1-s2'] = data['s1'].sub(data['s2'])
data
```

```python
data['s1/s2'] = data['s1'].div(data['s2'])
data
```

```python
data['s1xs2'] = data['s1'] * data['s2']
data
```

```python
#polynomial feature
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3, interaction_only=False, include_bias=True)

#input: s1, s2 
d = poly.fit_transform(data[['s1','s2']].values)

#check output name
poly.get_feature_names(['s1','s2'])
```

```python
pd.DataFrame(d, columns = poly.get_feature_names(['s1','s2']))
```

```python
#number to categorical value
data['diabetes_bin']=pd.qcut(data.diabetes, 3, labels=['dia_low','dia_medium','dia_high'])
data['bmi_bin']=pd.qcut(data.diabetes, 4, labels=['low','mean','medium','high'])
data['age_bin']=pd.qcut(data.diabetes, 3, labels=['young','adult','old'])
data
```

```python
cat = pd.Categorical(data.bmi_bin, 
                     categories=['low', 'mean', 
                                 'medium', 'high'], 
                     ordered=True)
cat
```

```python
labels, unique = pd.factorize(cat, sort=True)
data['bmi_ordinal'] = labels
```

```python
data
```

```python
#one hot encoding
from sklearn.preprocessing import OneHotEncoder

onehot = OneHotEncoder(dtype=np.int, sparse=True)
nominals = pd.DataFrame(
    onehot.fit_transform(data[['diabetes_bin', 'age_bin']])\
    .toarray(),
    columns=['dia_low', 'dia_medium', 'dia_high', 'young','adult', 'old'])

nominals['bmi_ordinal'] = data.bmi_ordinal
nominals
```

```python
#Discretization
from sklearn.preprocessing import KBinsDiscretizer  
est = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
X_ordinal = est.fit_transform(data.diabetes.values.reshape(-1,1)) 

data['diabetes_ordinal']=X_ordinal
data
```

```python
#binarization
from sklearn.preprocessing import Binarizer 

# For age, let threshold be 0.01
binarizer_1 = Binarizer(0.01)
data['age_binarizer']=binarizer_1.fit_transform(data.age.values.reshape(-1,1))
data
```

```python
#custom transformation
from sklearn.preprocessing import FunctionTransformer

transformer = FunctionTransformer(np.log1p, validate=True)
d = transformer.fit_transform(data.s1.values.reshape(-1, 1))

#the same
#data.s1.apply(lambda x : np.log1p(x))

data['custom_log1p']=d
data
```
