```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
house = load_boston()
data = pd.DataFrame(house.data,
                    columns=house.feature_names)
```

```python
#StandardScalerscaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
d=scaler.fit_transform(data.B.values.reshape(-1, 1))

data['standard_s']=d
data
```

```python
#min-max sacle
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-3,3))
d=scaler.fit_transform(data.B.values.reshape(-1, 1))

data['min_max_s']=d
data
```

```python
#max-abs sacle
from sklearn.preprocessing import MaxAbsScaler

scaler = MaxAbsScaler()
d=scaler.fit_transform(data.B.values.reshape(-1, 1))

data['max_abs_s']=d
data
```

```python
#robost sacle
from sklearn.preprocessing import RobustScaler

robust = RobustScaler(quantile_range = (0.1,0.9))
d=robust.fit_transform(data.B.values.reshape(-1, 1))

data['robost_s']=d
data
```

```python
#max norm
data['B_norm_max']=data.B/data.B.max()
data
```

```python
#norm_l1
data['B_norm_l1']=data.B/data.B.sum()
data
```

```python
#norm l2

data['B_norm_l2']=data.B/sum(data.B**2)
data
```
