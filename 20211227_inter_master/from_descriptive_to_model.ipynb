```python
#check quality of type of wine
red_wine=df.loc[df['type']=='red','quality']
white_wine=df.loc[df['type']=='white','quality']


sns.set_style("dark")
sns.distplot(red_wine, norm_hist=True, kde=False, color="red", label="Red wine")
sns.distplot(white_wine, norm_hist=True, kde=False, color="white", label="White wine")
sns.utils.axlabel("Quality score", "Density")
plt.legend()
plt.show()
```

# use t-test to check the quality of red/white wine is in significant difference or not
# the result tells us the quality of white wine is better than red wine (tstat: 9.686, and p-value: 0)

```python
print("STD:", df.groupby(['type'])[['quality']].agg(['std']))

tstat, pvalue, df1=sm.stats.ttest_ind(white_wine,red_wine)
print('tstat: %.3f p-value: %.4f' % (tstat, pvalue))
```

```python
#just sampling 
def take_sample(df,size,replace=False):
    return df.loc[np.random.choice(df.index, size=size, replace=replace)]
```

```python
#make samples amount are the same 
reds_sample=take_sample(df.loc[df['type']=='red'],200,False)
white_sample=take_sample(df.loc[df['type']=='white'],200,False)

df_sample=pd.concat([reds_sample,white_sample])
```

```python
#flag the sample data from original data set
df['in_sample']=np.where(df.index.isin(df_sample.index),1.,0.)
```

```python
df.loc[ (df.in_sample==1.0) & (df.type=='white')]
```

```python
pd.crosstab(df.in_sample, df.type, margins=all)
```

```python
sns.set_style("darkgrid")
g=sns.pairplot(df_sample, kind='reg', plot_kws={"ci":False, "x_jitter":0.25, "y_jitter":0.25}, hue='type', diag_kind='hist',\
            diag_kws={"bins":10, "alpha":1.0}, palette=dict(red="red", white="black"), markers=["o","s"], vars=['quality','alcohol','residual_sugar'])

plt.suptitle("Histograms and Scatter plots of quality, alcohol, residual sugar", fontsize=14, horizontalalignment='center', verticalalignment='top', x=0.5, y=0.999)
plt.show()
```

```python
g=sns.PairGrid(df_sample, hue='type', height=2)

def f(x, **kwargs):
    kwargs.pop("color")
    col = next(plt.gca()._get_lines.prop_cycler)['color']
    sns.kdeplot(x, color=col, **kwargs)

g.map_diag(f)
g.map_offdiag(plt.scatter)
g.add_legend()
plt.show()
```

```python
#OLS 
formula='quality ~ alcohol + chlorides + citric_acid + density + fixed_acidity + free_sulfur_dioxide + pH + residual_sugar + sulphates + total_sulfur_dioxide + volatile_acidity'
lm=ols(formula, data=df).fit()
lm.summary()
```

```python
print('Coefficients:',lm.params)
print('Coefficients Std Errors:',lm.bse)
print('R-Squared:',lm.rsquared_adj)
print('F-statistic:',lm.fvalue)
print('P-value:',lm.f_pvalue)
print('Number of obs:', lm.nobs)
print('Number of fitted values:',len(lm.fittedvalues))
```
