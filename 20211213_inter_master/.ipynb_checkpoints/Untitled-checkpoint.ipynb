```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
heatmap=sns.heatmap(df.corr(), cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
import numpy as np
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))
heatmap = sns.heatmap(df.corr(), mask=mask, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

```python
heatmap = sns.heatmap(df.corr()[['quality']].sort_values(by='quality', ascending=False), annot=True, cmap='BrBG')
```

```python
histogram = df['quality'].plot.hist()
plt.title('quality')
plt.show()
```

```python
df.plot(x ='quality', y='alcohol', kind = 'scatter')
plt.title('Scatter quality')
plt.show()
```

```python
dd=df.groupby('quality')['alcohol'].mean()
data={'quality': dd.index.tolist(),
     'alcohol': dd.values.tolist()
    }
df1=pd.DataFrame(data)
df1.plot(x ='quality', y='alcohol', kind = 'line')
plt.title('Line-chart quality/ mean of alcohol')
plt.show()
```

```python
#multi-line chart 
x = [7, 14, 21, 28, 35, 42, 49]
y = [5, 12, 19, 21, 31, 27, 35]
z = [3, 5, 11, 20, 15, 29, 31]

plt.plot(x, y, 'g', marker=r'o', label='Line y', linestyle='--')
plt.plot(x, z, 'r', marker=r's', label='Line z', linestyle=':')


plt.legend()
plt.show()
```

```python
df1.plot(x ='quality', y='alcohol', kind = 'bar')
plt.title('Bar-chart quality/ mean of alcohol')
plt.show()
```

```python
df2=df1.set_index('quality')
df2.plot.pie(y='alcohol',figsize=(5, 5), autopct='%1.1f%%', startangle=90)
plt.title('Pie-chart quality/ mean of alcohol')
plt.show()
```

```python
dd=df.quality.value_counts()
data={'quality_total': dd.values.tolist()}
df2=pd.DataFrame(data, index=dd.index)

df2.plot.pie(y='quality_total',figsize=(10, 10), autopct='%1.1f%%', startangle=90)
plt.title('Pie-chart quality_total')
plt.show()
```

```python
boxplot = df.boxplot(column=['alcohol'])
boxplot.plot()
plt.title('Boxplot alcohol')
plt.show()
```

```python
df1=df[['quality','residual sugar','fixed acidity']]
df1=df1.iloc[0:100]
sns.boxplot(x="variable", y="value", data=pd.melt(df1))
plt.title('Boxplot multi-columns')
plt.show()
```
